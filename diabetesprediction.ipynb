```
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```
diabetes_df=pd.read_csv('http://bit.ly/ADiabetesDataset')
diabetes_df.head()
```

```
data_x= diabetes_df.iloc[:,[0,1,2,4,5,7]].values
data_x #independent variable
```

```
data_y=diabetes_df.iloc[:,-1].values
data_y.head() #dependent variable
```

```
from sklearn.model_selection import train_test_split
X_train,x_test, Y_train, y_test=train_test_split(data_x, data_y,test_size=0.30, random_state=25)
print("X train", X_train.shape)
print("Y train", Y_train.shape)
print("X test", x_test.shape)
print("Y test:", y_test.shape)
```

```
from sklearn.preprocessing import StandardScaler 
sc=StandardScaler()
X_train= sc.fit_transform(X_train)
x_test= sc.transform(x_test)
```

```
from sklearn.naive_bayes import GaussianNB
naive_c=GaussianNB()
naive_c.fit(X_train, Y_train)
```

```
naive_y_prediction=naive_c.predict(x_test)
```

```
from sklearn.metrics import classification_report, accuracy_score
print("Accuracy score:", accuracy_score(naive_y_prediction, y_test))
print("Classification report", classification_report(y_test, naive_y_prediction))
```
